```python
# Load and preprocess your dataset using ImageDataGenerator >>aumaugmenting image data
train_datagen = ImageDataGenerator (
    rescale=1./255, 
    shear_range=0.2,
    #  zoom_range=0.2,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    vertical_flip=True,  # Add vertical flipping
    brightness_range=[0.8, 1.2],  # Adjust brightness
    horizontal_flip=True
    )


# train1_datagen = ImageDataGenerator (
#      rescale=1./255, 
#     shear_range=0.2,
#     #  zoom_range=0.2,
#     rotation_range=20,
#     width_shift_range=0.2,
#     height_shift_range=0.2,
#     horizontal_flip=True
#     )



test_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)



# train_datagen = ImageDataGenerator(
#     rescale=1./255,
#     shear_range=0.2,
#     # zoom_range=0.2,
#     rotation_range=20,
#     width_shift_range=0.2,
#     height_shift_range=0.2,
#     # vertical_flip=True,  # Add vertical flipping
#     brightness_range=[0.8, 1.2] , # Adjust brightness
#     horizontal_flip=True,
#     validation_split=0.2) # set validation split




# specified subsets for training and validation, which are commented out as well.
train_generator = train_datagen.flow_from_directory('D:/amm/adv_data/train', target_size=(48, 48), batch_size=16, class_mode='categorical',color_mode='grayscale',
                                                    # subset='training'
                                                    )
# train1_generator = train1_datagen.flow_from_directory('D:/amm/data/train/1', target_size=(48, 48), batch_size=128, class_mode='categorical',color_mode='grayscale',subset='training')
test_generator = test_datagen.flow_from_directory('D:/amm/adv_data/test', target_size=(48, 48), batch_size=16,color_mode='grayscale', class_mode='categorical')
validation_generator = train_datagen.flow_from_directory(
    'D:/amm/adv_data/val',
    target_size=(48, 48),
    color_mode='grayscale',
    batch_size=16,
    class_mode='categorical',
    # subset='validation'
    )


# D:/amm/data/train
# D:/amm/data/test
```

```python
model = tf.keras.Sequential([
   tf.keras.layers.Conv2D(16, kernel_size=(3,3), activation='relu', input_shape=(48,48,1)),
   tf.keras.layers.MaxPool2D(pool_size=(2, 2)),   #bt2ll noise  >> returns the maximum value
   tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'),
   tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
   #  tf.keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu'),
   #  tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
   tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
   tf.keras.layers.MaxPool2D(pool_size=(2, 2)),
   tf.keras.layers.Flatten(),   #Flattens the input, preparing it for the fully connected layers.  1d array
   
   # relu  (btsbly el mohm)   where neurons can become inactive during training and always output zero.
   # replaces all negative pixel values in the feature map by zero.
   tf.keras.layers.Dense(64, activation='relu'),  #Fully connected layer with 64 neurons (hidden)
   tf.keras.layers.Dropout(rate=0.5),  # decrease neurons

   tf.keras.layers.Dense(7, activation='softmax')   #3dd classes>>output layer 
])
```

```python
# Adam optimizer, loss function will be categorical cross entropy, evaluation metric will be accuracy
opt = keras.optimizers.Adam(learning_rate=0.001)  #0.0001  rate determines the step size at each iteration while moving toward a minimum of the loss function
# Compile the model
model.compile(optimizer= opt, loss='categorical_crossentropy', metrics=['accuracy'])   #optmize learning rate 

# A list of metrics to evaluate the model's performance during training
```

```python
model.summary()
```

```python
history = model.fit(train_generator, validation_data=validation_generator, epochs=99,
                    batch_size=16,callbacks=[EarlyStopping(monitor='val_loss', patience=7, verbose=0)])
# (monitor='val_loss'). If the validation loss does not improve for a certain number of epochs
# verbose=0 parameter means that there will be no output about the early stopping process.
```

```python
test_loss, test_acc = model.evaluate(test_generator)
print(f'Test Accuracy: {test_acc}, Test loss: {test_loss}')
```

```python
# Save the trained model
model.save('D:/amm/data/train')
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(accuracy) + 1)
#Train and validation accuracy
plt.plot(epochs, accuracy, 'b', label='Training accurarcy')
plt.plot(epochs, val_accuracy, 'r', label='Validation accurarcy')
plt.title('Training and Validation accurarcy')
plt.legend()

plt.figure() # to plot new figure
#Train and validation loss
plt.plot(epochs, loss, 'b', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and Validation loss')
plt.legend()
plt.show()
```

```python
# Load your trained model
model = tf.keras.models.load_model('D:/amm/data/train')

# Load and preprocess the image (adjust the path accordingly)
img_path = 'D:/amm/adv_data/val/disgust/S071_006_00000012 - Copy - Copy.png'
img = image.load_img(img_path, target_size=(48, 48), color_mode='grayscale')
img_array = image.img_to_array(img)   #This line converts the image (loaded using image.load_img()) to a NumPy array. 
img_array = np.expand_dims(img_array, axis=0)   #This line adds an extra dimension to the array to make it compatible with the input shape expected by the model
img_array = img_array / 255.0  # Normalize pixel values

# Make predictions
predictions = model.predict(img_array)

# Map the class indices to emotion labels
emotion_labels = ['angry', 'contempt', 'disgust', 'fear', 'happy', 'sad', 'suprise']

# Display the image
plt.imshow(img)
plt.axis('off')
plt.show()

# Display the predicted emotion scores
plt.figure(figsize=(8, 4))
plt.bar(emotion_labels, predictions[0], color='skyblue')
plt.title('Predicted Emotion Scores')
plt.xlabel('Emotion')
plt.ylabel('Score')
plt.show()

# Display the predicted emotion label
predicted_class_index = np.argmax(predictions)
predicted_emotion = emotion_labels[predicted_class_index]
print(f"Predicted Emotion: {predicted_emotion}")
```
